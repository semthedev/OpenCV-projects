## Basic Image Operations



```python
# Importing the necessary libraries
import cv2  # OpenCV for computer vision tasks
import numpy as np  # NumPy for numerical computations
import matplotlib.pyplot as plt  # Matplotlib for plotting and data visualization
```

```python
def show_image(image, title="Image"):
    """
    Function for displaying an image in Google Colab.
    """
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display
    plt.figure(figsize=(6,6))
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
path = '/content/drive/MyDrive/8 семестр весна (4 курс 2025 года)/Система компьютерного зрения /lab-01-basic-image-processing/example1.jpg'

# image upload
picture = cv2.imread(path)
show_image(picture, "Original image")
```

```python
# image cropping
cropped = picture[10:400, 43:130]
show_image(cropped, "Cropped image")
```

```python
# image resizing
picture = cv2.imread(path)
d = (int(picture.shape[1] * 100 / 50), int(picture.shape[0] * 100 / 50))
resized = cv2.resize(picture, d, interpolation=cv2.INTER_AREA)
show_image(resized, "Resized image")
```

```python
# image rotation
(h, w, d) = picture.shape
c = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(c, 45, 1)
rotated = cv2.warpAffine(picture, M, (w, h))
show_image(rotated, "Rotated image")
```

```python
o = picture.copy() # сopy image
cv2.line(o, (0, 0), (w, h), (0, 0, 255), 2) # draw diagonal lines
cv2.line(o, (0, h), (w, 0), (0, 0, 255), 2)
cv2.putText(o, "this rare animal is restricted", (30, 30), cv2.FONT_HERSHEY_TRIPLEX, 0.8, (0, 255, 0), 1) # add text
cv2.putText(o, "to their prime habitat", (60, 60), cv2.FONT_HERSHEY_TRIPLEX, 0.8, (0, 255, 0), 1)
show_image(o) # show image
```
