```python
# Importing the necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image in Google Colab.
def show_image(image, title="Image"):
    """
    Function for displaying an image in Google Colab.
    """
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct display
    plt.figure(figsize=(6,6))
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
# Paths to the images.
path1 = '/content/drive/MyDrive/8 семестр весна (4 курс 2025 года)/Система компьютерного зрения /lab-02/image_1.jpg'
path2 = '/content/drive/MyDrive/8 семестр весна (4 курс 2025 года)/Система компьютерного зрения /lab-02/image_2.jpg'
path3 = '/content/drive/MyDrive/8 семестр весна (4 курс 2025 года)/Система компьютерного зрения /lab-02/image_3.jpg'
```

```python
# Read images from the specified paths.
first_image = cv2.imread(path1)
second_image = cv2.imread(path2)
third_image = cv2.imread(path3)

# Display all images.
show = [first_image, second_image, third_image]
for i in show:
    show_image(i)
```

```python
# Function to enhance the image using a median filter.
def improve_image_with_median_filter(image_for_median):
  processed_image = cv2.medianBlur(image_for_median, 7)
  show_image(processed_image, "Медианный фильтр")
  cv2.imwrite('median_filter_image.jpg', processed_image)

improve_image_with_median_filter(first_image)
```

```python
# Function to enhance the image using a mean (average) filter.
def improve_image_with_mean_filter(image_for_mean):
  kernel = np.ones((5, 5), np.float32) / 25
  processed_image = cv2.filter2D(image_for_mean, -1, kernel)
  show_image(processed_image, "Средний фильтр")
  show_image(processed_image, "Средний фильтр")
  cv2.imwrite('mean_filter_image.jpg', processed_image)

improve_image_with_mean_filter(second_image)
```

```python
# Function to enhance the image using erosion.
def improve_image_with_erode(image_for_erode):
  kernel = np.ones((5, 5), np.uint8)
  processed_image = cv2.erode(image_for_erode, kernel, iterations=1)
  show_image(processed_image, "Erosion")
  cv2.imwrite('erode_processed_image.png', processed_image)

improve_image_with_erode(third_image)
```

```python
src = cv2.imread(path1, cv2.IMREAD_COLOR)
```

```python
# Apply Gaussian Blur to the image.
src_blurred = cv2.GaussianBlur(src, (3, 3), 0)
show_image(src_blurred, "Image after Gaussian Blur")
```

```python
# Convert the blurred image to grayscale.
src_gray = cv2.cvtColor(src_blurred, cv2.COLOR_BGR2GRAY)
show_image(src_gray, "Converted image to grayscale")
```

```python
# Settings for computing the Laplacian: image depth and kernel size.
ddepth = cv2.CV_16S
kernel_size = 3

# Compute the Laplacian to detect edges.
laplacian = cv2.Laplacian(src_gray, ddepth, ksize=kernel_size)

# Convert the Laplacian result to absolute values for proper display.
abs_dst = cv2.convertScaleAbs(laplacian)
show_image(abs_dst, "Laplacian")
```
